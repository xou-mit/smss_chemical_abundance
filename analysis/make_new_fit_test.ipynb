```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

# Import normalized spectrum
wl = np.genfromtxt('/Users/xwou/Desktop/test',usecols=0)
fl = np.genfromtxt('/Users/xwou/Desktop/test',usecols=1)
```

```python
fl
```

```python
hdr = fits.Header()
hdr['RA'] = '00000'
hdr['DEC'] = "11111"
hdr['Tele'] = 'Magellan'
hdr['Inst'] = 'MIKE'
```

```python
# Create the column objects
c1 = fits.Column(name='wavelength', array=wl, format = 'E')
c2 = fits.Column(name='norm_flux', array=fl, format = 'E')
print(c1,c2)
t = fits.BinTableHDU.from_columns([c1, c2],header=hdr)
```

```python
t.writeto('/Users/xwou/Desktop/test_fit.fits',overwrite = True)
```
