```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import glob
import pandas as pd

matplotlib.rc('xtick', labelsize=24) 
matplotlib.rc('ytick', labelsize=24) 
```

```python
# Import the data
df = pd.read_csv('../data/td_lit/c_sc_co_all.csv')
```

```python
df
```

```python
# Plot the abundances
f = plt.figure(figsize=[30,10])
plt.scatter(pd.to_numeric(df['Fe/H'],errors='coerce'),pd.to_numeric(df['C/Fe'],errors='coerce'),
            s=400,marker='.',c='k')
plt.xlim([-5.1,-0.2])
plt.ylim([-0.81,1.1])
plt.xlabel('[Fe/H]',fontsize=24)
plt.ylabel('[C/Fe]',fontsize=24)
plt.show()
```

```python
# Plot the abundances
f = plt.figure(figsize=[30,10])
plt.scatter(pd.to_numeric(df['Fe/H'],errors='coerce'),pd.to_numeric(df['Sc/Fe'],errors='coerce'),
            s=400,marker='.',c='k')
plt.xlim([-5.1,-0.2])
plt.ylim([-0.81,1.1])
plt.xlabel('[Fe/H]',fontsize=24)
plt.ylabel('[Sc/Fe]',fontsize=24)
plt.show()
```
